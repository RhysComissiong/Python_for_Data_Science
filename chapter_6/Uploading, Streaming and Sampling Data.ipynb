```python
import os
data_dir = "../Data" # must include the ..
name = "/Colors.txt"
data_path = data_dir + name
print(data_path)
```

```python
# read data from txt file
with open(os.path.join(data_path),'r') as open_file:
    print("Colors.txt content:\n" + open_file.read())
```

```python
# stream data
with open(os.path.join(data_path),'r') as open_file:
    for observation in open_file:
        print('Reading Data: ' + observation)
```

```python
# image data
import matplotlib.image as img
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = img.imread("free_image.jpg")
print(image.shape) # horizontal pixels x vertical pixels x pixel depth
print(image.size)
plt.imshow(image)
plt.show()
```

```python
# sample the color.txt file
n = 2 # sample every second record
with open(data_path) as open_file:
    for j, observation in enumerate(open_file):
        if j % n==0:
            print('Reading Line: ' + str(j) + ' Content: ' + observation)
```

```python
# random sampling
from random import random
sample_size = 0.25
with open(data_path,'r') as open_file:
    for j, observation in enumerate(open_file):
        if random()<=sample_size:
            print('Reading Line: ' + str(j) + ' Content: ' + observation)
```

# Unstructured Data - picture = unstructured file

```python
from skimage.io import imread
from skimage.transform import resize
from matplotlib import pyplot as plt
import matplotlib.cm as cm

example_file = ("https://upload.wikimedia.org/" + "wikipedia/commons/7/7d/Dog_face.png")
image = imread(example_file, as_gray=True)
plt.imshow(image, cmap=cm.gray)
plt.show()
```

```python
print("data type: %s, shape: %s" % (type(image), image.shape))
```

```python
image2 = image[5:70,0:70]
plt.imshow(image2, cmap=cm.gray)
plt.show()
```

```python
image3 = resize(image2, (30,30), mode='symmetric')
plt.imshow(image3, cmap=cm.gray)
print("data type: %s, shape: %s" % (type(image3), image3.shape))
```

```python
image_row = image3.flatten()
print("data type: %s, shape: %s" % (type(image_row), image_row.shape))
```
